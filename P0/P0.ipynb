```python
# pip install pillow
from PIL import Image
# pip install opencv-python
import cv2

# pip install scikit-image
import imageio.v3 as iio
import matplotlib.pyplot as plt

def show_with_PIL(image_path):
  # leer una imagen con PIL
  image = Image.open(image_path)

  print(image.format)
  print(image.size)
  print(image.mode)

  # plt.imshow(image, cmap='gray')
  # plt.show()
  image.show()
  
def show_with_opencv(image_path):
  # leemos la image, imread(path, modo)
  # 0 escala de grisis
  # 1 color, solo rgb
  # -1 todos los canales, incluyendo el canal alfa si est disponible.
  img_grayscale = cv2.imread(image_path, 0)
  cv2.imshow('grayscale image',img_grayscale)
  cv2.waitKey(0)                     # Esperar hasta que se presione una tecla
  cv2.destroyAllWindows()            # Cerrar todas las ventanas abiertas

def show_with_scikit_image(image_path):
  image = iio.imread(uri=image_path)
  fig, ax = plt.subplots()
  ax.imshow(image)
  
def show_with_scikit_matplotlib(image_path):
  image = plt.imread(image_path)
  plt.imshow(image)  
  
show_with_PIL('./static/house.bmp')
show_with_opencv('./static/house.bmp')
show_with_scikit_image('./static/house.bmp')
show_with_scikit_matplotlib('./static/house.bmp')
```
